```python
# IBM Statistics for Data Science with Python - Module 7 Assignment
# Author: Fahmi Ahmed
# Compatible with JupyterLite (offline)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create sample dataset
data = {
    'carat': [0.23, 0.45, 0.71, 0.34, 0.29, 0.90, 1.02, 0.75, 0.30, 0.60],
    'price': [326, 945, 3100, 630, 520, 5000, 6100, 2900, 450, 1800],
    'cut': ['Ideal', 'Premium', 'Good', 'Very Good', 'Fair', 'Ideal', 'Good', 'Premium', 'Very Good', 'Fair'],
    'clarity': ['SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'SI1', 'SI2', 'VS2', 'VS1', 'VVS1']
}

df = pd.DataFrame(data)

# Show the dataset
print("Sample dataset:")
print(df)

# Histogram of prices
plt.figure()
plt.hist(df['price'], bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Boxplot: Carat by Cut
plt.figure()
cuts = df['cut'].unique()
cut_values = [df[df['cut'] == cut]['carat'] for cut in cuts]
plt.boxplot(cut_values, labels=cuts)
plt.title("Carat by Cut")
plt.xlabel("Cut")
plt.ylabel("Carat")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Scatter plot: Carat vs Price colored by Cut
plt.figure()
colors = {'Ideal': 'blue', 'Premium': 'green', 'Good': 'red', 'Very Good': 'purple', 'Fair': 'orange'}
for cut in df['cut'].unique():
    subset = df[df['cut'] == cut]
    plt.scatter(subset['carat'], subset['price'], label=cut, color=colors[cut])
plt.title("Carat vs Price by Cut")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Correlation matrix
plt.figure()
correlation = df[['carat', 'price']].corr()
plt.imshow(correlation, cmap='coolwarm', interpolation='none')
plt.colorbar(label="Correlation Coefficient")
plt.xticks([0, 1], ['carat', 'price'])
plt.yticks([0, 1], ['carat', 'price'])
for (i, j), val in np.ndenumerate(correlation.values):
    plt.text(j, i, f'{val:.2f}', ha='center', va='center', color='black')
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()

# Grouped mean price by clarity
grouped = df.groupby('clarity')['price'].mean()
print("\nMean price by clarity:")
print(grouped)

# Bar plot: Mean price by Clarity
plt.figure()
plt.bar(grouped.index, grouped.values, color='lightgreen', edgecolor='black')
plt.title("Mean Price by Clarity")
plt.xlabel("Clarity")
plt.ylabel("Mean Price")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
